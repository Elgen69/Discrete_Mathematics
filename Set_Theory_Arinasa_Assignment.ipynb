# Set Identities
---
1. Identity Laws:
- A $\cup$ $\emptyset$ = A
- A $\cap$ $\textbf{U}$ = A
<br>
2. Dominant Laws:
- A $\cup$ $\textbf{U}$ = $\textbf{U}$
- A $\cap$ $\emptyset$ = $\emptyset$
<br>
3. Idempotent Laws:
- A $\cup$ A = A
- A $\cap$ A = A
<br>
4. Double Negation Law or the Involution Law:
- $(A^{c})^{c}$ = $ \neg $($ \neg $ A) = A, where "$\neg$" is read as "logical not".
5. Commutative Laws:
- A $\cup$ B = B $\cup$ A
- A $\cap$ B = B $\cap$ A
<br>
6. Associative Laws:
- A $\cup$ (B $\cup$ C) = (A $\cup $ B) $ \cup $ C
- A $\cap$ (B $\cap$ C) = (A $\cap $ B) $ \cap $ C
<br>
7. Distributive Laws:
- A $\cap$ (B $\cup$ C) = (A $\cap $ B) $ \cup $ (A $\cap $ B)
- A $\cup$ (B $\cap$ C) = (A $\cup $ B) $ \cap $ (A $\cup $ B)
<br>
8. De Morgan's Law:
- $\neg$(A $\cup$ C) = ($ \neg $ A $\cap $ $ \neg$ b) = $(A \cup C)^{c}$ = ($A^{c}$ $\cap$ $B^{c}$)
- $\neg$(A $\cap$ C) = ($ \neg $ A $\cup $ $ \neg$ b) = $(A \cap C)^{c}$ = ($A^{c}$ $\cup$ $B^{c}$)
<br>
9. Absorption Laws:
- A $\cup$ (A $\cap$ B) = A
- A $\cap$ (B $\cup$ C) = A
<br>
10. Complement Laws:
- A $\cup$ $\neg$ A = A $\cup$ $A^{c}$ =A
- A $\cap$ $\neg$ A = A $\cap$ $A^{c}$ =$\emptyset$
- $\neg$ $\emptyset$ = $\emptyset^{c}$ = $\textbf{U}$
- $\neg$ $\textbf{U}$ = $\textbf{U}^{c}$ = =$\emptyset$
<br>

$\textit{Solution:}$

 Let $A$ and $B$ be sets. We want to show that $A \setminus B = A \cap B^C$. The identity $A \setminus B = A \cap B^C$ can be proven using the following set theory laws: 
 
<br>Commutative law of intersection: A $\cap$ B = B $\cap$ A 
<br> Distributive law of intersection over union:  A $\cap$ (B $\cup$ C) = (A $\cap $ B) $ \cup $ (A $\cap $ B)
<br> Complement law: $A $ $\cap $ $ A^C $ = ϕ (the empty set)$ 

$\underline{\textit{Other types of proof:}}$

The proof using contradiction is as follows:


Assume that $A \setminus B \neq A \cap B^C$. Then there exists an element $x$ such that $x \in A \setminus B$ and $x \notin A \cap B^C$.

This means that $x \in A$ and $x \notin B$, but $x \in B^C$. This is a contradiction, because $B^C$ is the set of elements that are not in $B$.

Therefore, our assumption must be false, and $A \setminus B = A \cap B^C$.


$\underline{\textit{The proof using element proving is as follows:}}$

Let $x \in A \setminus B$. Then $x \in A$ and $x \notin B$.

Since $x \in A$, we have $x \in A \cap B^C$.

Therefore, every element of $A \setminus B$ is also an element of $A \cap B^C$.

Conversely, let $x \in A \cap B^C$. Then $x \in A$ and $x \notin B$.

Since $x \notin B$, we have $x \in A \setminus B$.

Therefore, every element of $A \cap B^C$ is also an element of $A \setminus B$.

Hence, $A \setminus B = A \cap B^C$.

$\underline{\textit{The proof using example is as follows:}}$

Let $A = \{1, 2, 3\}$ and $B = \{2, 3\}$. Then $A \setminus B = \{1\}$ and $A \cap B^C = \{1\}$.

Therefore, $A \setminus B = A \cap B^C$.

**6.) |A $\cup$ B| = |A|+|B| - |A $\cap$ B|**

$\underline{\textit{The proof by contradiction:}}$

Assume that $|A \cup B| \neq |A|+|B| - |A \cap B|$. Then there exists an integer $n$ such that $|A \cup B| = n$ and $|A|+|B| - |A \cap B| = n+1$.

This means that there are $n$ elements in $A \cup B$ and $n+1$ elements in $A \cup B$. But this is a contradiction, because $A \cup B$ cannot have more elements than $A \cup B$.

Therefore, our assumption must be false, and $|A \cup B| = |A|+|B| - |A \cap B|$.

$\underline{\textit{The proof by element proving:}}$

Let $x \in A \cup B$. Then $x \in A$ or $x \in B$.

If $x \in A$, then $x$ is counted once in the cardinality of $A$ and once in the cardinality of $A \cup B$.

$|A \cup B| = |A| + |B| - |A \cap B|$

If $x \in B$, then $x$ is counted once in the cardinality of $B$ and once in the cardinality of $A \cup B$.

Therefore, every element of $A \cup B$ is counted exactly once in the sum $|A|+|B|$.

Conversely, let $x$ be an element that is counted exactly once in the sum $|A|+|B|$. Then $x$ must be in either $A$ or $B$.

Therefore, $x$ is in $A \cup B$.

Hence, $|A \cup B| = |A|+|B| - |A \cap B|$.

$\underline{\textit{The proof by example:}}$

Let $A = \{1, 2, 3\}$ and $B = \{2, 3, 4\}$. Then $|A \cup B| = 5$, $|A| = 3$, and $|B| = 3$.

Therefore, $|A \cup B| = |A|+|B| - |A \cap B|$.

$\underline{\textit{The proof by conditional proving:}}$

Let $x \in A \cup B$. Then $x \in A$ or $x \in B$.

Conditionally, assume that $x \notin A \cup B$. Then $x \notin A$ and $x \notin B$.

$x \notin A \cup B \implies x \notin A \land x \notin B$

This means that $x \notin A$ and $x \notin B$. But this contradicts the fact that $x \in A \cup B$.

Therefore, our assumption must be false, and $x \in A \cup B$.

Hence, $|A \cup B| = |A|+|B| - |A \cap B|$.

# 4. Finite Sets and The Counting Principles

In a Survey of 120 People, it was found that:
- 65 read Newsweek Magazine
- 45 read Times Magazine
- 42 read Fortune Magazine
- 20 read Both Newsweek and Time
- 25 read both Newsweek and Fortune
- 15 read both Time and Fortune
- 8 read all three magazines

```python
from matplotlib_venn import venn3, venn3_circles
import matplotlib.pyplot as plt

# This code creates a Venn diagram of magazine readership.

# Given:
#   - The number of people who read only Newsweek.
#   - The number of people who read only Time.
#   - The number of people who read only Fortune.
#   - The number of people who read Newsweek and Time.
#   - The number of people who read Newsweek and Fortune.
#   - The number of people who read Time and Fortune.
#   - The number of people who read all three magazines.

n_only = 65 - (20 + 25 - 8)
t_only = 45 - (20 + 15 - 8)
f_only = 42 - (25 + 15 - 8)
n_and_t = 20 - 8
n_and_f = 25 - 8
t_and_f = 15 - 8
all_three = 8

# Let X be the number of people who read at least one of the three magazines.
X = 65 + 45 + 42 - 20 - 25 - 15 + 8

# Let Y be the total number of people who have read exactly one magazine.
f_readers = 42 - 25 - 15 + 8
n_readers = 65 - 25 - 20 + 8
t_readers = 45 - 15 -20 + 8
Y = f_readers + n_readers + t_readers

# Create the Venn diagram.
venn = venn3(subsets=(n_only, t_only, n_and_t,
                      f_only, n_and_f, t_and_f, all_three),
             set_labels=('Newsweek', 'Time', 'Fortune'))

# Draw circles for the universal set.
venn_circles = venn3_circles(subsets=(n_only, t_only, n_and_t,
                                      f_only, n_and_f, t_and_f, all_three))
venn_circles[0].set_lw(2.0)  # Set line width for circle 1
venn_circles[1].set_lw(2.0)  # Set line width for circle 2
venn_circles[2].set_lw(2.0)  # Set line width for circle 3

# Print the number of people who read at least one of the 3 magazines and the total number of people
# who read exactly one magazine.
print("Number of people who read at least one of the 3 magazines:", X)
print("Total number of people who read at exactly one magazine:", Y)

plt.title("Venn Diagram of Magazine Readership")
plt.show()



```
